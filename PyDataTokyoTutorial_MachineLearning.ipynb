### ライブラリのインポート

```python
# ライブラリをインポートします
import numpy as np
import pandas as pd
import pydot, StringIO
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from IPython.core.display import Image
from sklearn.preprocessing import LabelEncoder
from sklearn.cross_validation import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import cross_val_score, KFold, StratifiedKFold
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import roc_curve, auc

plt.style.use('ggplot')
```

```python
# データを読み込みます
# trainは正解付きのトレーニングデータ、testは正解なしのテストデータです
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
df_train.tail()
```

```python
df_train.info()
```

```python
# 年齢の欠損値を平均値とします
df_train['Age'].fillna(df_train['Age'].mean(),inplace=True)
df_test['Age'].fillna(df_train['Age'].mean(),inplace=True)
```

```python
# scikit-learnは文字情報を扱えないため、性別を[0, 1]に変換します
le = LabelEncoder()
le.fit(df_train['Sex'])
df_train['Sex'] = le.transform(df_train['Sex'])
df_test['Sex'] = le.transform(df_test['Sex'])

# どのように変化されたか確認します
# 0: female, 1:male
print le.classes_
```

```python
# データを確認します
df_train[['Sex', 'Age']].tail()
```

## ジェンダーモデルによる生存者推定、推定値の評価

```python
# 必要なデータを選びます
X = df_train['Sex']
y = df_train['Survived']
X_test = df_test['Sex']
```

```python
# データを学習用と評価用に分割します
VAL_SIZE = 0.2
RANDOM_STATE = 0
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=VAL_SIZE, random_state=RANDOM_STATE)
```

```python
# 女性(0)は生存(1)、男性(1)は死亡(0)となるように計算します
y_train_pred = X_train * -1 + 1
y_val_pred = X_val * -1 + 1
```

```python
# 評価のための関数を定義します
def evaluate(y_train, y_train_pred, y_val, y_val_pred):
    print "Accuracy on Training Set: {0:.3f}".format(accuracy_score(y_train, y_train_pred))
    print
    print "Accuracy on Validation Set: {0:.3f}".format(accuracy_score(y_val, y_val_pred))
    print 
    print "Classification Report on Validation Set:"
    print classification_report(y_val, y_val_pred)
    print
    print "Confusion Matrix on Validation Set:"
    print confusion_matrix(y_val, y_val_pred)

    fig, ax = plt.subplots()
    cm = plt.cm.get_cmap('binary')
    mat = ax.matshow(confusion_matrix(y_val, y_val_pred),cmap=cm)
    fig.colorbar(mat)
    ax.set_title('Confusion matrix')
    ax.set_ylabel('True label')
    ax.set_xlabel('Predicted label')
```

```python
# モデルを評価します
evaluate(y_train, y_train_pred, y_val, y_val_pred)
```

```python
# テストデータについても同様に推測します
y_test_pred = X_test * -1 + 1

# Kaggleにアップロードするデータを作成します
# (Kaggleのスコアは0.76555になります)
df_kaggle = pd.DataFrame({'PassengerId': df_test['PassengerId'], 'Survived':np.array(y_test_pred)})
df_kaggle.to_csv('kaggle_gendermodel.csv', index=False)
```

## ロジスティック回帰、クロスバリデーション

```python
# 特徴量を増やしてみましょう
X = df_train[['Age', 'Pclass', 'Sex']]
y = df_train['Survived']
X_test = df_train[['Age', 'Pclass', 'Sex']]
y_test = df_train['Survived']
```

```python
# 新しい特徴量を作成し、特徴量を2つにします
X['Pclass_Sex'] = X['Pclass'] + X['Sex']
X.drop(['Pclass', 'Sex'], axis=1, inplace=True)
X_test['Pclass_Sex'] = X_test['Pclass'] + X_test['Sex']
X_test.drop(['Pclass', 'Sex'], axis=1, inplace=True)
```

```python
# データを学習用と評価用に分割します
VAL_SIZE = 0.2
RANDOM_STATE = 0
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=VAL_SIZE, random_state=RANDOM_STATE)
```

```python
# データを可視化します
# 1等に乗船した若い女性ほど、生存率が高いことが分かります
xmin, xmax = -5, 85
ymin, ymax = 0.5, 4.5
labels = ['Survived', 'Dead']

fig, ax = plt.subplots()
cm = plt.cm.RdBu
cm_bright = ListedColormap(['#FF0000', '#0000FF'])
ax.scatter(X_train[:, 0], X_train[:, 1]+(np.random.rand(len(X_train))-0.5)*0.1, c=y_train, cmap=cm_bright)
ax.scatter(X_val[:, 0], X_val[:, 1]+(np.random.rand(len(X_val))-0.5)*0.1, c=y_val, cmap=cm_bright, alpha=0.5)
ax.set_xlabel('Age')
ax.set_ylabel('Pclass_Sex')
ax.set_xlim(xmin, xmax)
ax.set_ylim(ymin, ymax)
ax.legend()
```

```python
# 分類器(Classifier)を定義し、学習します
clf = LogisticRegression()
clf.fit(X_train, y_train)
```

```python
# 推測します
y_train_pred = clf.predict(X_train)
y_val_pred = clf.predict(X_val)
```

```python
# モデルを評価します
evaluate(y_train, y_train_pred, y_val, y_val_pred)
```

```python
# 分類の境界を可視化します

h = 0.02
xmin, xmax = -5, 85
ymin, ymax = 0.5, 4.5
xx, yy = np.meshgrid(np.arange(xmin, xmax, h), np.arange(ymin, ymax, h))
Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]
Z = Z.reshape(xx.shape)

fig, ax = plt.subplots()
cm = plt.cm.RdBu
levels = np.linspace(0, 1.0, 5)
cm_bright = ListedColormap(['#FF0000', '#0000FF'])
contour = ax.contourf(xx, yy, Z, cmap=cm, levels=levels, alpha=0.8)
ax.scatter(X_train[:, 0], X_train[:, 1]+(np.random.rand(len(X_train))-0.5)*0.1, c=y_train, cmap=cm_bright)
ax.scatter(X_val[:, 0], X_val[:, 1]+(np.random.rand(len(X_val))-0.5)*0.1, c=y_val, cmap=cm_bright, alpha=0.5)
ax.set_xlabel('Age')
ax.set_ylabel('Pclass_Sex')
ax.set_xlim(xmin, xmax)
ax.set_ylim(ymin, ymax)
fig.colorbar(contour)

x1 = xmin
x2 = xmax
y1 = -1*(clf.intercept_[0]+clf.coef_[0][0]*xmin)/clf.coef_[0][1]
y2 = -1*(clf.intercept_[0]+clf.coef_[0][0]*xmax)/clf.coef_[0][1]
ax.plot([x1, x2] ,[y1, y2], 'k--')
```

```python
# クロスバリデーションの関数を定義します
RANDOM_STATE = 0
def cross_val(clf, X, y, K):
    cv = KFold(len(y), K, shuffle=True, random_state=RANDOM_STATE)
    scores = cross_val_score(clf, X, y, cv=cv)
    print 'Scores:', scores
    print ("Mean Score: {0:.3f} (+/-{1:.3f})").format(scores.mean(), scores.std() * 2)
```

## グリッドサーチ

```python
clf = DecisionTreeClassifier()
param_grid = {'max_depth': [2, 3, 4, 5], 'min_samples_leaf': [2, 3, 4, 5]}
```

```python
grid_search = GridSearchCV(clf, param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(X, y)
```

```python
print 'Scores: {:.3f}'.format(grid_search.score(X_test, y_test))
print 'Best Parameter Choice:', grid_search.best_params_
```

```python
scores = [x[1] for x in grid_search.grid_scores_]
scores = np.array(scores).reshape(4, 4)

fig, ax = plt.subplots()
cm = plt.cm.jet
mat = ax.matshow(scores, cmap=cm)
ax.set_xlabel('max_depth')
ax.set_ylabel('min_samples_leaf')
ax.set_xticklabels(['']+param_grid['max_depth'])
ax.set_yticklabels(['']+param_grid['min_samples_leaf'])
fig.colorbar(mat)
```
