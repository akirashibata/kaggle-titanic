```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

## データの読み込み＆確認

```python
fn_train = 'train.csv'
fn_test = 'test.csv'
df_train = pd.read_csv(fn_train)
df_test = pd.read_csv(fn_test)
```

## データのクリーニング

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.cross_validation import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.svm import SVC
# from sklearn.cross_validation import cross_val_score, KFold
# from sklearn.grid_search import GridSearchCV

TEST_SIZE = 0.2
RANDOM_STATE = 0
```

```python
df_train['Age'].fillna(df_train['Age'].mean(),inplace=True)
```

```python
le = LabelEncoder()
le.fit(df_train['Sex'])
print le.classes_
df_train['Sex'] = le.transform(df_train['Sex'])
```

```python
df_train.tail()
```

## ジェンダーモデル

```python
X = df_train['Sex']
y = df_train['Survived']
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE)
```

```python
y_train_pred = X_train * -1 + 1
y_val_pred = X_val * -1 + 1
```

```python
def evaluate(y, y_pred):
    print "Accuracy: {0:.3f}".format(accuracy_score(y, y_pred))
    print 
    print "Classification Report:"
    print classification_report(y, y_pred)
    print
    print "Confusion Matrix:"
    print confusion_matrix(y, y_pred)

    fig, ax = plt.subplots()
    cm = plt.cm.get_cmap('binary')
    mat = ax.matshow(confusion_matrix(y, y_pred),cmap=cm)
    fig.colorbar(mat)
    ax.set_title('Confusion matrix')
    ax.set_ylabel('True label')
    ax.set_xlabel('Predicted label')
```

```python
evaluate(y_train, y_train_pred)
```

```python
evaluate(y_val, y_val_pred)
```

## SGDClassifier

```python
X = df_train[['Age', 'Pclass']]
y = df_train['Survived']
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE)
```

```python
clf = SGDClassifier(n_iter=1000)
clf.fit(X_train, y_train)
print clf
print 'coef_:', clf.coef_[0]
print 'intercept_:', clf.intercept_[0]
# 18.284312925 + -0.67690055 * x + -5.56803025 * y = 0
# y = -1 * (18.284312925 + -0.67690055 * x) / -5.56803025
```

```python
xmin, xmax = -5, 85
ymin, ymax = 0.5, 3.5

fig, ax = plt.subplots()
cm = plt.cm.RdBu
cm_bright = ListedColormap(['#FF0000', '#0000FF'])
ax.scatter(X_train[:, 0], X_train[:, 1]+(np.random.rand(len(X_train))-0.5)*0.1, c=y_train, cmap=cm_bright)
ax.scatter(X_val[:, 0], X_val[:, 1]+(np.random.rand(len(X_val))-0.5)*0.1, c=y_val, cmap=cm_bright, alpha=0.5)
ax.set_xlabel('Age')
ax.set_ylabel('Pclass')
ax.set_xlim(xmin, xmax)
ax.set_ylim(ymin, ymax)

x1 = xmin
x2 = xmax
y1 = -1*(clf.intercept_[0]+clf.coef_[0][0]*x1)/clf.coef_[0][1]
y2 = -1*(clf.intercept_[0]+clf.coef_[0][0]*x2)/clf.coef_[0][1]
ax.plot([x1, x2] ,[y1, y2], 'k--')
```

```python
y_train_pred = clf.predict(X_train)
y_val_pred = clf.predict(X_val)
```

```python
evaluate(y_val, y_val_pred)
```

## LogisticRegression

```python
clf = LogisticRegression()
clf.fit(X_train, y_train)
```

```python
h = 0.02
xmin, xmax = -5, 85
ymin, ymax = 0.5, 4.5
xx, yy = np.meshgrid(np.arange(xmin, xmax, h), np.arange(ymin, ymax, h))

if hasattr(clf, "decision_function"):
    Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
else:
    Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]
Z = Z.reshape(xx.shape)

fig, ax = plt.subplots()
cm = plt.cm.RdBu
cm_bright = ListedColormap(['#FF0000', '#0000FF'])
ax.contourf(xx, yy, Z, cmap=cm, alpha=.8)
ax.scatter(X_train[:, 0], X_train[:, 1]+(np.random.rand(len(X_train))-0.5)*0.1, c=y_train, cmap=cm_bright)
ax.scatter(X_val[:, 0], X_val[:, 1]+(np.random.rand(len(X_val))-0.5)*0.1, c=y_val, cmap=cm_bright, alpha=0.5)
ax.set_xlabel('Age')
ax.set_ylabel('Pclass')
ax.set_xlim(xmin, xmax)
ax.set_ylim(ymin, ymax)
```

```python
y_train_pred = clf.predict(X_train)
y_val_pred = clf.predict(X_val)
```

```python
evaluate(y_val, y_val_pred)
```
